# 性能度量
- 均方误差 mean square error
$$
E(f;D)=\frac{1}{m}\sum_{i=1}^{m}(f(x_i)-y_i)^2
$$
$$
E(f;\mathcal{D})=\int_{x \sim \mathcal{D}}(f(x)-y)^2p(x)dx
$$

- 错误率 
$$
E(f:D)=\frac{1}{m}\sum_{i=1}^m\mathbb{I}(f(x_i) \neq y_i)
$$
更一般的针对数据分布和概率密度,错误率有
$$
E(f;D)=\int_{x \sim \mathcal{D}} \mathbb{I}(f(x) \neq y)p(x)dx
$$

- 精度
$$
E(f:D)=\frac{1}{m}\sum_{i=1}^m\mathbb{I}(f(x_i) = y_i)
$$
更一般的针对数据分布和概率密度,精度有
$$
E(f;D)=\int_{x \sim \mathcal{D}} \mathbb{I}(f(x) = y)p(x)dx
$$

## 查准率、查全率和F1

分类结果混淆矩阵 | 预测正例 |预测反例
---|---|---
实际正例 | TP | FN
实际反例 | FP | TN

查准率P
$$
P=\frac{TP}{TP+FP}
$$
查全率R
$$
R = \frac{TP}{TP+FN}
$$

PR曲线与平衡点BEP

F1
$$
F1=\frac{2\times P \times R}{P+R}

\frac{2}{F}= \frac{1}{P} + \frac{1}{R}
$$
更一般的有
$$
\frac{1}{F_\beta}=\frac{1}{1+\beta ^2} (\frac{1}{P} +\frac{\beta ^2}{R})
$$

多次训练和测试得到多个混淆矩阵
$$
macro-P=\frac{1}{n}\sum_{i=1}^nP_i
$$

ROC的全称是“受试者工作特征”（Receiver Operating Characteristic）,纵轴是真正例率TPR，横轴是假正例率FPR
$$
TPR= \frac{TP}{TP+FN}

FPR= \frac{FP}{FP+TN}
$$

AUC为ROC曲线下的面积

代价敏感错误率与代价曲线
$$
E(f;D;cost)=\frac{1}{m}(\sum_{x_i \in D^+} \mathbb{I}(f(x_i) \neq y_i) \times cost_{01} + \sum_{x_i \in D^-} \mathbb{I}(f(x_i) \neq y_i) \times cost_{10})
$$


# 比较校验
## 假设校验
- 一次留出法估计：二项检验 binomial test
- 多次留出法估计：t校验 t-test

## 交叉验证t校验
对学习器A和B，使用k折交叉验证法,计算差值的均值和方差，做t校验

$$
\Delta_i=\epsilon_i^A - \epsilon_i^B
\tau_t = |\frac{\sqrt{k} \mu}{\sigma}|
$$

5次2折交叉验证

## McNemar校验
两学习器分类差别关联表
$$
\tau_{\chi^2}=\frac{(|e_{01} - e_{10}|-1)^2}{e_{01}+e_{10}}
$$

## Friedman检验与Nemenyi后续检验
在N个数据集上比较k个算法，r表示算法的平均序值
$$
\bar{r}=\frac{\sum_{i=1}^k r_i}{k}=\frac{k+1}{2}
var(r)=\frac{k^2-1}{12}
$$
后续检验（post-hoc testy)
$$
CD=q_\alpha \sqrt{\frac{k(k+1)}{6N}}
$$
